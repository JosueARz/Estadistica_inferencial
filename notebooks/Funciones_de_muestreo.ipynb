## Muestreo sistemático

```python
# Creamos una función que seguira una regla para elección de datos.

def sistematic(data, steps):
    """
    Hacemos una seleción de datos estableciendo una regla, steps.
    data : DataFrame 
    steps : int 

    return : DataFrame
    """
    indexes = np.arange(0,len(data), step=steps)
    systematic_sample = data.iloc[indexes]
    return systematic_sample
```

```python
# a la funcion sistematic le pasamos nuestra base de datos o dataframe y cada cuento queremos que seleccione un dato
muestreo_sistematico = sistematic(econdata, 4)
```

## Muestreo estratificado

```python
# creamos una nueva variable dentro de nuestro DataFrame
# vamos a crearlo uniendo la delegacion y el tipo

econdata['estratificado'] = econdata['delegacion'] + ',' + econdata['tipo']

```

Queremos que los datos tengan las siguentes proporciones:

1) Hoteles en Cuauhtémoc 0.5

2) Museos en Cuauhtémoc 0.2

3) Hoteles Venustiano Carranza 0.1

4) Mercados en Cuauhtémoc 0.1

5) Mercados en Venustiano Carranza 0.1

```python
def data_estratificada(data, nombres_columnas_estrat, valores_estrat, prop_estrat, random_state=None):
    
    df_estrat = pd.DataFrame(columns = data.columns) # Creamos un data frame vacío con los nombres de las columnas de econdata

    pos = -1
    for i in range(len(valores_estrat)): # iteración sobre los valores estratificados
        pos += 1
        if pos == len(valores_estrat) - 1: 
            ratio_len = len(data) - len(df_estrat) # si es la iteración final calcula el número de valores de salida tenga el mismo número de filas que de entrada
        else:
            ratio_len = int(len(data) * prop_estrat[i]) # calcula el número de filas según la proporción deseada

        df_filtrado = data[data[nombres_columnas_estrat] ==valores_estrat[i]] # filtra los datos de origen según los valores seleccionados en la estratificación de datos
        df_temp = df_filtrado.sample(replace=True, n=ratio_len, random_state=random_state) # haz un sample de los datos filtrados usando la ratio que hemos calculado
        
        df_estrat = pd.concat([df_estrat, df_temp]) # junta las tablas de sample con la estratificada para producir el resultado final
        
    return df_estrat
```

```python
valores_estrat = ['Cuautémoc,Hotel', 'Cuautémoc,Museo', 'Venustiano Carranza,Hotel', 'Cuauhtémoc,Mercado','Venustiano Carranza,Mercado']
prop_estrat = [0.5, 0.2, 0.1, 0.1,0.1]
df_estrat = data_estratificada(econdata, 'estratificado', valores_estrat, prop_estrat, random_state=42)
```
